```python
import kagglehub
import shutil
import os
from pathlib import Path


def get_fraud_data():
    """
    Descarga el dataset de Credit Card Fraud y lo mueve a una carpeta local.
    """
    print("🚀 Iniciando descarga desde Kaggle...")

    # 1. Descarga a la caché temporal de kagglehub
    temp_path = kagglehub.dataset_download("mlg-ulb/creditcardfraud")

    # 2. Definir ruta de destino en tu proyecto
    dest_folder = Path("../data")
    dest_folder.mkdir(exist_ok=True)

    # 3. Buscar el CSV y moverlo a la carpeta /data
    csv_filename = "creditcard.csv"
    source_file = Path(temp_path) / csv_filename
    final_path = dest_folder / csv_filename

    if source_file.exists():
        shutil.copy(source_file, final_path)
        print(f"✅ Dataset guardado exitosamente en: {final_path}")
        return str(final_path)
    else:
        # En caso de que el nombre varíe, buscamos cualquier .csv
        for file in os.listdir(temp_path):
            if file.endswith(".csv"):
                shutil.copy(Path(temp_path) / file, dest_folder / file)
                print(f"✅ Dataset ({file}) guardado en: {dest_folder / file}")
                return str(dest_folder / file)

    print("❌ No se encontró ningún archivo CSV en la descarga.")
    return None


if __name__ == "__main__":
    path = get_fraud_data()
    if path:
        print(f"\nListo para procesar. Ubicación: {os.path.abspath(path)}")
```
